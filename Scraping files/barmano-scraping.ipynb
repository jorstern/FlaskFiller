```python
def getSoup(url):
    html = requests.get(url)
    soup = BS(html.content, "html.parser")
    return soup
```

```python
soup1 = getSoup("http://www.barmano.com/drinks/recipe-alcohol-list.html")
link_boxes = soup1.findAll("li", {"class": "big"})
box1 = soup1.findAll("li", {"class": "big"})[0]
```

```python
category_links = []
for box in link_boxes:
    category = box.next
    link = category['href']
    category_links.append(link)
```

```python
moresoup = getSoup(category_links[0])
moresoup1 = moresoup.findAll("div", {"class": "itemInfoContainer"})
moresoup1[0].find()
```

```python
morelinks = []
for cat in category_links:
    categ_soup = getSoup(cat)
    drink_hyper = categ_soup.findAll("div", {"class": "itemInfoContainer"})
    #alcoholic range category if I want to label them better later
    #print(drink_hyper)
    for d in drink_hyper:
        drink = d.find()
        drink_link = drink['href']
        morelinks.append(drink_link)

morelinks
        
```

```python
soup = getSoup(morelinks[0])
ingredient_tag = soup.findAll("span", {"class": "ingredient"})
print(soup.find("h1").get_text())
ingredient_tag[0].find("a")["title"]
```

```python
cocktail_dict = {}
for dr in morelinks:
    soup = getSoup(dr)
    ingredient_tag = soup.findAll("span", {"class": "ingredient"})
    ingredient_list = []
    for ing in ingredient_tag:
        ingredient = ing.find("a")["title"]
        ingredient_list.append(ingredient)
    #print(ingredient_list)
    drink_name_tag = soup.find("h1")
    drink_name = drink_name_tag.get_text()
    print(drink_name)
    cocktail_dict[drink_name] = ingredient_list
```

```python
print(len(cocktail_dict))
```

```python
import json
data = json.dumps(cocktail_dict, indent=4, separators=(',', ': '))
print(type(data))
print(data)
with open("cocktails_barmano.json","w") as f:
  f.write(data)
```
