```python
def getSoup(url):
    html = requests.get(url)
    soup = BS(html.content, "html.parser")
    return soup
```

```python
links = soup.findAll('a', href=True) #Finds all 'a' tags with an href object (i.e. all hyperlinks)
```

```python
links
```

```python
#link_boxes = soup.findAll('div', attrs={'class' : "overlay"}, href=True)
#link_boxes = soup.findAll('a', class="overlay", href=True)
#link_boxes = soup.findAll('div', class="overlay", href=True)
link_boxes = soup.findAll("a", {"class": "overlay"}, href=True)
```

```python
link_boxes
```

```python
cocktail_categories = []
for l in link_boxes:
    if "mosaic" in l['href']:
        if "http://liquor.com/" in l['href']:
            cocktail_categories.append(l['href'])
        else:
            cocktail_categories.append("http://liquor.com" + l['href'])
```

```python
cocktail_categories
```

```python
#from time import sleep
```

```python
all_cocktail_links = []
for category_link in cocktail_categories:
    drink_soup = getSoup(category_link)
    drink_boxes = drink_soup.findAll("a", {"target": "_blank", "class": "overlay"}, href=True)
    #print(drink_boxes)
    for d in drink_boxes:
        if "recipes" in d['href']:
            if "http://liquor.com" in d['href']:
                all_cocktail_links.append(d['href'])
            else:
                all_cocktail_links.append("http://liquor.com" + d['href'])
```

```python
all_cocktail_links
```

```python
print(len(all_cocktail_links))
```

```python
cocktail_dict = {}

for cock in all_cocktail_links:
    soup_drink = getSoup(cock) #opens the html for each drink
    ingredients = soup_drink.find_all("div", {"class" : "col-xs-8 x-recipe-ingredient"})
    ingredients_cleaned = []
    for item in ingredients:
        text = item.get_text().split()
        text = ' '.join(text)
        ingredients_cleaned.append(text)
    drink_name_tag = soup_drink.find("h1", itemprop="name")
    drink_name = drink_name_tag.get_text()
    cocktail_dict[drink_name] = ingredients_cleaned
```

```python
cleaned_inv = {}
for drink in cocktail_dict:
    for ing in cocktail_dict[drink]:
        i = ing.lower()
        if "vodka" in i:
            i = "vodka"
        elif "rum" in i:
            i = "rum"
        elif "bourbon" in i:
            i = "bourbon"
        elif "tequila" in i:
            i = "tequila"
        elif "liqueur" in i:
            i = "liqueur"
        elif "baileys" in i:
            i = "baileys"
        elif "gin" in i:
            i = "gin"
        elif "whiskey" in i:
            i = "whiskey"
        elif "wine" in i:
            i = "wine"
        if i in cleaned_inv.keys():
            cleaned_inv[i].append(drink)
        else:
            cleaned_inv[i] = [drink]
print (cleaned_inv)
```

```python
from operator import itemgetter

cleaned_ingredient_counts = {}
for ingredient in cleaned_inv:
    length = len(cleaned_inv[ingredient])
    if length > 0: 
        cleaned_ingredient_counts[ingredient] = length
ings = sorted(cleaned_ingredient_counts.items(), key=itemgetter(1), reverse=True)
top_alcs = []
alcs = ["vodka", "rum", "bourbon", "tequila", "liqueur", "baileys", "gin", "whiskey", "wine"]
for pair in ings:
    if pair[0] in alcs:
        top_alcs.append(pair)
        
```
